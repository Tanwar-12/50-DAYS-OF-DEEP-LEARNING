# **PERCEPTRON LOSS FUNCTION**

 * The loss function is typically used to measure the error between the predicted output and the actual output.
* The goal of training a perceptron is to minimize this error, and the loss function helps quantify how far off the predictions are from the true values.

* For a perceptron, which is a type of binary classifier, a common loss function is the "perceptron loss" or "0-1 loss" function. In this loss function, the error is simply 0 if the prediction matches the true label and 1 otherwise.

_**IMPORTING THE  LIBRARIES**_

```python
from sklearn.datasets import make_classification
import numpy as np
import matplotlib.pyplot as plt
```

_**LOAD THE DATA**_

```python
X, y = make_classification(n_samples=100, n_features=2, n_informative=1,n_redundant=0,
                           n_classes=2, n_clusters_per_class=1, random_state=41,hypercube=False,class_sep=15)
```

_**PLOTTING POINTS**_

```python
plt.figure(figsize=(7,4))
plt.scatter(X[:,0],X[:,1],c=y,cmap='coolwarm',s=40)
plt.show()
```

```python
def perceptron(X, y):
    # Initialize weights and bias
    w1 = w2 = b = 1
    # Learning rate
    lr = 0.1
    
    # Iterate through a fixed number of epochs (here, 1000)
    for j in range(1000):
        # Iterate through each data point
        for i in range(X.shape[0]):
            # Compute the dot product of weights and input features plus bias
            z = w1 * X[i][0] + w2 * X[i][1] + b

            # Check if the point is misclassified
            if z * y[i] < 0:
                # Update weights and bias using the Perceptron update rule
                w1 = w1 + lr * y[i] * X[i][0]
                w2 = w2 + lr * y[i] * X[i][1]
                b = b + lr * y[i]
            
    return w1, w2, b
```

```python
w1,w2,b = perceptron(X,y)
```

```python
m = -(w1/w2)
c = -(b/w2)

print(m,c)
```

```python
x_input = np.linspace(-3,3,100)
y_input = m*x_input + c
```

**_PLOTTING GRAPH_**

```python
plt.figure(figsize=(7,4))
plt.plot(x_input, y_input, color='red', linewidth=3)
plt.scatter(X[:,0], X[:,1], c=y, cmap='coolwarm', s=40)

# Define the regions
plt.fill_between(x_input, y_input, min(y_input), where=(y_input >= min(y_input)), color='red', alpha=0.3)
plt.fill_between(x_input, y_input, 2, where=(y_input <= 2), color='green', alpha=0.3)

plt.ylim(-3, 2)
plt.show()
```
