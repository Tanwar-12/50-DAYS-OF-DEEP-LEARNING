# **MULTILAYER PERCEPTRON**

**A Multilayer perceptron (MLP) is a type of artificial neural network that consists of multiple layers of nodes (neurons), with each layer fully connected to the next layer. The architecture typically includes an input layer, one or more hidden layers, and an output layer.**

### **Here's a breakdown of its components:**


**Input Layer**: 
* The first layer of the MLP, where the input data is fed into the network. Each node in this layer represents a feature of the input data.

**Hidden Layers**: 
* These are intermediate layers between the input and output layers. Each hidden layer consists of multiple nodes, and each node is connected to every node in the previous and subsequent layers. These layers are where the network learns complex patterns and features from the input data through a process called forward propagation.

**Output Layer**:
* The final layer of the MLP, which produces the network's output. The number of nodes in this layer depends on the nature of the problem - for example, a binary classification task might have one output node for each class, while a regression task might have a single output node.

_**MLPs use activation functions, such as the sigmoid function, hyperbolic tangent (tanh) function, or rectified linear unit (ReLU) function, to introduce non-linearity into the network, enabling it to learn complex relationships in the data**_

## _**PRACTICAL IMPLEMENTATION**_ 
_**IMPORTING LIBARIES**_

```python
import numpy as np
import pandas as p
```

_**LOAD THE DATASET**_

```python
df=pd.read_csv('Churn_Modelling.csv')
```

```python
df.head()
```

```python
df.drop(columns=['RowNumber','CustomerId','Surname'],inplace=True)
```

```python
df.head()
```

```python
df['Geography'].value_counts()
```

```python
df['Gender'].value_counts()
```

```python
df=pd.get_dummies(df,columns=['Geography','Gender'],drop_first=True)
```

```python
df.head()
```

```python
X=df.drop(columns='Exited')
y=df['Exited'].values
```

```python
from sklearn.model_selection import train_test_split
X_train,X_test,y_train,y_test=train_test_split(X,y,test_size=0.2,random_state=0)
```

```python
from sklearn.preprocessing import StandardScaler
scaler=StandardScaler()

X_train_trf=scaler.fit_transform(X_train)
X_test_trf=scaler.transform(X_test)
```

```python
import tensorflow
from tensorflow import keras
from tensorflow.keras import Sequential
from tensorflow.keras.layers import Dense
```

```python
model=Sequential()

model.add(Dense(11,activation='sigmoid',input_dim=11))
model.add(Dense(11,activation='sigmoid'))
model.add(Dense(1,activation='sigmoid'))
```

```python
model.summary()
```

```python
model.compile(optimizer='Adam',loss='binary_crossentropy',metrics=['accuracy'])
```

```python
history=model.fit(X_train,y_train,batch_size=50,epochs=100,verbose=1,validation_split=0.2)
```

```python
y_pred=model.predict(X_test)
```

```python
y_pred
```

```python
y_pred=y_pred.argmax(axis=-1)
```

```python
from sklearn.metrics import accuracy_score
accuracy_score(y_test,y_pred)
```

```python
import matplotlib.pyplot as plt

plt.plot(history.history['loss'])
plt.plot(history.history['val_loss'])
```

```python
plt.plot(history.history['accuracy'])
plt.plot(history.history['val_accuracy'])
```
